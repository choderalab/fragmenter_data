```python
%matplotlib inline
import matplotlib.pyplot as plt
from fragmenter import chemi
from openeye import oechem, oequacpac
import oenotebook as oenb
import json
import numpy as np
import glob
import cmiles
import os
from matplotlib.backends.backend_pdf import PdfPages
#import seaborn as sbn
import pandas as pd
import matplotlib.image as mpimg
from cclib.parser import Psi4
from cclib.parser.utils import convertor
```

```python
imatinib_psi4 = glob.glob('../conjugation/bond_order_without_geomopt/Imatinib/Imatinib_*.txt')
```

```python
mols = []
tagged_smiles = '[H:41][c:7]1[c:8]([c:3]([c:4]([n:5][c:6]1[H:40])[H:39])[c:2]2[c:1]([c:37]([n:36][c:10]([n:9]2)[N:11]([H:43])[c:12]3[c:17]([c:16]([c:15]([c:14]([c:13]3[C:35]([H:65])([H:66])[H:67])[H:44])[H:45])[N:18]([H:47])[C:19](=[O:34])[c:20]4[c:25]([c:24]([c:23]([c:22]([c:21]4[H:48])[H:49])[C:26]([H:52])([H:53])[N:27]5[C:32]([C:31]([N:30]([C:29]([C:28]5([H:54])[H:55])([H:56])[H:57])[C:33]([H:62])([H:63])[H:64])([H:58])[H:59])([H:60])[H:61])[H:50])[H:51])[H:46])[H:68])[H:38])[H:42]'
oemol = cmiles.utils.load_molecule(tagged_smiles)
atom_map = cmiles.utils.get_atom_map(molecule=oemol, mapped_smiles=tagged_smiles)
connectivity = cmiles.utils.get_connectivity_table(oemol, atom_map)
for im in imatinib_psi4:
    log = Psi4(im)
    data = log.parse()
    geometry = data.converged_geometries[0].flatten()*cmiles.utils.ANGSROM_2_BOHR
    symbols = []
    inverse_map = cmiles.utils.invert_atom_map(cmiles.utils._symbols)
    for i in data.atomnos:
        symbols.append(inverse_map[i])
    json_mol = {'symbols': symbols,
           'geometry': list(geometry),
           'connectivity': connectivity}
    mols.append(cmiles.utils.mol_from_json(json_mol))
    
```

```python
bond_orders = {}
for m in mols:
    # Calculate WBO
    m_copy = oechem.OEMol(m)
    #am1charges = oequacpac.OEAM1Charges(optimize=False)
    if oequacpac.OEAssignPartialCharges(m_copy, oequacpac.OECharges_AM1BCCSym):
    # Save WBO
        for bond in m_copy.GetBonds():
            bond_key = (bond.GetBgn().GetIdx(), bond.GetEnd().GetIdx())
            if bond_key not in bond_orders:
                bond_orders[bond_key] = []
            bond_orders[bond_key].append(bond.GetData('WibergBondOrder'))
```

```python
# remove hydrogen and seperate rings 
rings_bo = {}
other_bo = {}
for bond in oemol.GetBonds():
    a1 = bond.GetBgn()
    a2 = bond.GetEnd()
    key = (a1.GetMapIdx()-1, a2.GetMapIdx()-1)
    reverse_key = (key[-1], key[0])
    if bond.IsInRing():
        try:
            rings_bo[key] = bond_orders[key]
        except KeyError:
            rings_bo[reverse_key] = bond_orders[reverse_key]
    elif a1.IsHydrogen() or a2.IsHydrogen():
        continue
    else:
    
        try:
            other_bo[key] = bond_orders[key]
        except KeyError:
            other_bo[reverse_key] = bond_orders[reverse_key]
```

```python
import seaborn as sbn
fig, axes = plt.subplots(11, 1)
fig.dpi = 400
#plt.box(False)
for i, bond in enumerate(other_bo):
    ax = plt.subplot(11, 1, i+1)
    ax.spines['left'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    if i == 0:
        sbn.kdeplot(other_bo[bond], shade=True, label='openeye AM1')
        #sbn.kdeplot(bond_order_std_rot_bonds_unopt['Imatinib']['wiberg_bo'][bond], shade=True, label='non-optimized')
    else:
        sbn.kdeplot(other_bo[bond], shade=True)
        #sbn.kdeplot(bond_order_std_rot_bonds_unopt['Imatinib']['wiberg_bo'][bond], shade=True)
    plt.xlim(0.90, 1.27)
    plt.yticks([])
    #ax.set_yticklabels(bond_order_std_rot_bonds_opt['Imatinib']['bonds'], fontsize=6, rotation=0)
    #ax.yaxis.grid(False)
    ax.yaxis.set_label_coords(-0.05, 0)
    plt.ylabel((bond[0]+1, bond[1]+1), rotation=0, size=8)
    if i != 10:
        plt.xticks([])
    else:
        plt.xlabel('Bond order')
    if i == 0:
        plt.legend(prop={'size': 10}, bbox_to_anchor=(1.35, 1))
        plt.title('Distribution of OE WBO')
#overlap=0.01
#h_pad = 5 + (- 5*(1 + overlap))
#fig.tight_layout(h_pad=h_pad)
plt.savefig('imatinib_oewbo_dist.pdf', bbox_inches='tight')
```
